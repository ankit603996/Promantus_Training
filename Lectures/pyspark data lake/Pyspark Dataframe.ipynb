```
filelocation = "dbfs:/FileStore/tables/UsedCars.csv"

sparkDF = spark.read \
                .format('csv')\
                .option('inferSchema', 'true') \
                .option('header', 'true') \
                .load(filelocation)
```

# Filtering

```
import random
from datetime import datetime
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

```
sparkDF.filter(col('Age')> 50).count()
```

# Descriptive Stats

```
display(sparkDF.describe())
```

```
sparkDF.groupBy('FuelType').agg(max('Age'), min('Age'), avg('Age')).show()
```

```
sparkDF.groupBy('FuelType').agg(max('Price'), min('Price'), avg('Price')).show()
```

```
from pyspark.sql.functions import when
sparkDF = sparkDF.withColumn('FuelType', \
                  when(sparkDF['FuelType'] == 'diesel', 'Diesel').otherwise(sparkDF['FuelType']))
```

```
sparkDF.groupBy('FuelType').agg(max('Price'), min('Price'), avg('Price')).show()
```

## Program to traverse through the whole directory, inspect the given folder & get the files with given extension

```python
def traverse_get_files(search_directory, folder_lookup, file_extension_lookup):
    for root, dirs, files in os.walk(search_directory):
        if folder_lookup in root:
            for file in files:
                if file.endswith(file_extension_lookup):
                    print(root + '\\'+ file)
```

```python
traverse_get_files('D:\Learnings\Python_Training', 'Directories', '.ipynb')
```

```python
traverse_get_files('D:\Learnings\Python_Training', 'Directories', '.csv')
```

## Just in case, you want to store the output in the collection like "list"

```python
def traverse_get_files_output(search_directory, folder_lookup, file_extension_lookup):
    out_list = []
    for root, dirs, files in os.walk(search_directory):
        if folder_lookup in root:
            for file in files:
                if file.endswith(file_extension_lookup):
                    out_list.append(root + '\\'+ file)
    return out_list
```

```python
traverse_get_files_output('D:\Learnings\Python_Training', 'modules', '.py')
```

Now, we can see that there are several variables like `workclass`, `occupation` and `native_country` which contain missing values. Generally, the missing values are coded as `NaN` and python will detect them with the usual command of `df.isnull().sum()`.

But, in this case the missing values are coded as `?`. Python fail to detect these as missing values because it do not consider `?` as missing values. So, I have to replace `?` with `NaN` so that Python can detect these missing values.

Lets explore these variables and replace `?` with `NaN`.

```python
# check frequency distribution of values in workclass variable

df.workclass.value_counts()
```

```python
# replace '?' values in workclass variable with `NaN`


df['workclass'].replace('?', np.NaN, inplace=True)
```

```python
# again check the frequency distribution of values in workclass variable

df.workclass.value_counts()
```

# **Model training and Logging the Experiments**

```python
#Necessary imports for mlflow
import mlflow
import mlflow.pyfunc
import mlflow.sklearn
from mlflow.models.signature import infer_signature
from mlflow.utils.environment import _mlflow_conda_env
import os
# train a Gaussian Naive Bayes classifier on the training set
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import roc_auc_score
```

## Naive Bayes Experiment

```python
with mlflow.start_run(run_name = 'Naive Bayes'):
  # instantiate the model
  gnb = GaussianNB()
  # fit the model
  gnb.fit(X_train, y_train)
  
  # Make predictions
  predictions = gnb.predict(X_test) #Prediction on test dataset
  y_pred_train = gnb.predict(X_train) #Prediction on train dataset
  
  #Train accuracy
  train_accuracy = accuracy_score(y_train, y_pred_train)
  
  #Test accuracy
  test_accuracy = accuracy_score(y_test, predictions)
  #AUC Calculation
#   ROC_AUC = roc_auc_score(y_test, predictions)
   
  #Test Precision and Recall calculation
  cm = confusion_matrix(y_test, predictions)
  TP = cm[0,0]
  TN = cm[1,1]
  FP = cm[0,1]
  FN = cm[1,0]
  
  #Precision score
  precision = TP / float(TP + FP)
  
  #Recall
  recall = TP / float(TP + FN)
  
  # Log parameters to be tracked for comparisons
  mlflow.log_param("priors", gnb.get_params()['priors'])
  mlflow.log_param("var_smoothing", gnb.get_params()['var_smoothing'])
  
  #Infer schema and make it as signature
  signature = infer_signature(X_train, gnb.predict(X_train))

  # Log model
  mlflow.sklearn.log_model(gnb, "naive-bayes-model",  signature=signature)
  
    
  # Log metrics
  mlflow.log_metric("Training Accuracy", train_accuracy)
  mlflow.log_metric("Test Accuracy", test_accuracy)
  mlflow.log_metric("Precision", precision)
  mlflow.log_metric("Recall", recall)
```

## Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV

with mlflow.start_run(run_name = 'Random Forest Classifier'):
  # instantiate the model
  clf=RandomForestClassifier(n_estimators=100)
  # fit the model
  param_dist = {'max_depth': [2, 3, 4],
              'bootstrap': [True, False],
              'max_features': ['auto', 'sqrt', 'log2', None],
              'criterion': ['gini', 'entropy']
               }
  cv_rf = GridSearchCV(clf, cv = 10, param_grid=param_dist, n_jobs = 3)
  cv_rf.fit(X_train, y_train)
  
  # Make predictions
  predictions = cv_rf.predict(X_test) #Prediction on test dataset
  y_pred_train = cv_rf.predict(X_train) #Prediction on train dataset
  
  #Train accuracy
  train_accuracy = accuracy_score(y_train, y_pred_train)
  
  #Test accuracy
  test_accuracy = accuracy_score(y_test, predictions)
  #AUC Calculation
#   ROC_AUC = roc_auc_score(y_test, predictions)
   
  #Test Precision and Recall calculation
  cm = confusion_matrix(y_test, predictions)
  TP = cm[0,0]
  TN = cm[1,1]
  FP = cm[0,1]
  FN = cm[1,0]
  
  #Precision score
  precision = TP / float(TP + FP)
  
  #Recall
  recall = TP / float(TP + FN)
  
  # Log parameters to be tracked for comparisons
  mlflow.log_param("boostrap", cv_rf.best_params_['bootstrap'])
  mlflow.log_param("max_depth", cv_rf.best_params_['max_depth'])
  mlflow.log_param("max_features", cv_rf.best_params_['max_features'])
  mlflow.log_param("criterion", cv_rf.best_params_['criterion'])
  
  #Infer schema and make it as signature
  signature = infer_signature(X_train, cv_rf.predict(X_train))

  # Log model with signature of the schema of the model
  mlflow.sklearn.log_model(cv_rf, "random-forest-model",  signature=signature)
  
    
  # Log metrics
  mlflow.log_metric("Training Accuracy", train_accuracy)
  mlflow.log_metric("Test Accuracy", test_accuracy)
  mlflow.log_metric("Precision", precision)
  mlflow.log_metric("Recall", recall)
```

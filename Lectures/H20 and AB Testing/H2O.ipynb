### Train-Test Split

```python
#Prepare predictors and response columns
predictors = df.drop(1,axis=1)
response = df.col_names[1]
```

```python
predictors = predictors.col_names
```

```python
predictors
```

```python
df["is_canceled"]= df["is_canceled"].asfactor()
```

```python
## a little different syntax that scikit learn
splits = df.split_frame(ratios=[0.70], seed=1)  

train = splits[0]
test = splits[1]
```

```python
## get the shape of each of the datasets
print('train shape', train.shape)
print('test shape', test.shape)
```

## ML Model
### Gradient Boosting

```python
## Gradient Boosting Estimator with cross validation
gbm_model = H2OGradientBoostingEstimator( ntrees=50, max_depth=3, min_rows=2, learn_rate=0.2, nfolds=5)
```

```python
# train the model
gbm_model.train(predictors, response, training_frame=train, validation_frame=test)
```

```python
## accuracy, auc, confusion matrix
print('accuracy', gbm_model.accuracy())
print('auc', gbm_model.auc())
print('confusion matrix', gbm_model.confusion_matrix())
print(gbm_model.confusion_matrix(thresholds=[0.1, 0.5, 0.99]))
```

```python
performance_test = gbm_model.model_performance(test)
performance_test.confusion_matrix()
```

```python
predictions = gbm_model.predict(test)
predictions.head(5)
```

## ML Model
### Random Forest

```python
## Gradient Boosting Estimator with cross validation
rf_model = H2ORandomForestEstimator(ntrees=50, max_depth=3, min_rows=2, nfolds=5)
```

```python
# train the model
rf_model.train(predictors, response, training_frame=train, validation_frame=test)
```

```python
## accuracy, auc, confusion matrix
print('accuracy', rf_model.accuracy())
print('auc', rf_model.auc())
print('confusion matrix', rf_model.confusion_matrix())
print(rf_model.confusion_matrix(thresholds=[0.1, 0.5, 0.99]))
```

```python
test_performance = rf_model.model_performance(test)
```

```python
test_performance.confusion_matrix()
```

```python
predictions = rf_model.predict(test)
predictions.head(5)
```

## Grid Search

```python
## Grid Search
ntrees_opt = [5,50]
max_depth_opt = [2,3]
learn_rate_opt = [0.1,0.2]

hyper_params = {'ntrees': ntrees_opt, 
                'max_depth': max_depth_opt,
                'learn_rate': learn_rate_opt}
search_criteria = {'strategy': 'RandomDiscrete', 'max_runtime_secs': 60}  #updated
```

```python
gs = H2OGridSearch(H2OGradientBoostingEstimator, hyper_params = hyper_params, search_criteria = search_criteria)
```

```python
gs.train(predictors, response, training_frame=train, validation_frame=test)
```

```python
# sort the models by auc score
gbm_gridperf = gs.get_grid(sort_by='auc', decreasing=True)
```

```python
# get the best model
best_model = gbm_gridperf.models[0]
```

```python
# get the auc score
gbm_perf = best_model.model_performance(test)
print(gbm_perf.auc())
```

```python
class Promanthus:
    """
    Function to create the hiring candidates
    """
    
    # class Variable
    candidateCount = 0
    
    #special method. Python calls this function only when new instance is created for this Class
    def __init__(self, name, education, stream, college, current_residency):
        #These are instance variables
        self.name = name
        self.education = education
        self.stream = stream
        self.college = college
        self.current_residency = current_residency
        Promanthus.candidateCount += 1
    
    def displayEmployeeCount(self):
        """
        Method to show the number of candidates in Promanthus
        """
        print("Candidates count: {}".format(Promanthus.candidateCount))
        
        
    def showCandidateDetails(self):
        """
        Method to show the candidate details in Promanthus
        """
        print("Name: {}, Education: {}, Stream :{}, College: {}, Current Location: {}"\
              .format(self.name, self.education, self.stream, self.college, self.current_residency))   
    
```

```python
candidate1 = Promanthus("Yogalakshmi", "B.E", "ECE", "Anna University Trichy", "Chennai")
```

```python
candidate2 = Promanthus("Balaji", "MCA", "Computer Applications", "Anna University", "Chennai")
```

## Static Method

```python
class CompanyCandidate:
    """
    Function to create the hiring candidates
    """
    
    # class Variable
    candidateCount = 0
    company_name = "Promanthus"
    company_address = "Chennai"
    
    #special method. Python calls this function only when new instance is created for this Class
    def __init__(self, name, education, stream, college, current_residency):
        #These are instance variables
        self.name = name
        self.education = education
        self.stream = stream
        self.college = college
        self.current_residency = current_residency
        Promanthus.candidateCount += 1
    
    def displayEmployeeCount(self):
        """
        Method to show the number of candidates in Promanthus
        """
        print("Candidates count: {}".format(Promanthus.candidateCount))
        
        
    def showCandidateDetails(self):
        """
        Method to show the candidate details in Promanthus
        """
        print("Name: {}, Education: {}, Stream :{}, College: {}, Current Location: {}"\
              .format(self.name, self.education, self.stream, self.college, self.current_residency))   
    
    @classmethod
    def company_info(cls):
        """
        Method to show the company info
        """
        print("Company Name: {}".format(cls.company_name))
        print("Company Address: {}".format(cls.company_address))
        
    #Static method does not work with instance object as well as class object
    @staticmethod
    def general_info():
        """
        Method to show the general information of the class
        """
        print("""This class is created to store the Candidates hires of the Company.\n\nThe Company details can be seen by calling company_info method""")
```

```python
CompanyCandidate.general_info()
```

```python
CompanyCandidate.company_info()
```

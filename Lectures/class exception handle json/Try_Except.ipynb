## Exception Handling : ZeroDivisionError

```python
x = 9
y = 0

try:
    print(x/y)
except(ZeroDivisionError):
    print("The denominator cannot be zero")
```

### Check try block is executing well

```python
x = 9
y = 3

try:
    print(x/y)
except(ZeroDivisionError):
    print("The denominator cannot be zero")
```
